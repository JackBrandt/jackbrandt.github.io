Then we can begin joining one by one

```python
ascqr_cjr=pd.merge(ascqr,cjr,how='outer',on='Facility Name')
ascqr_cjr.head()
ascqr_cjr.to_csv('ascqr_cjr.csv', index=False)
# This join had zero overlap...
```

```python
# Fix Complications and Deaths format to be one listing per facility
complicationsdeaths['Score'] = complicationsdeaths['Score'].replace('Not Available', np.nan)
fixed_compls = complicationsdeaths.pivot_table(
    index=['Facility Name', 'City/Town', 'State', 'County/Parish'],
    columns='Measure Name',
    values='Score',
    aggfunc='first'  # use first score if duplicates exist
).reset_index()
fixed_compls.columns.name = None
```

```python
# Remove rows where most of the data is missing
measure_cols = fixed_compls.columns.difference(['Facility Name', 'City/Town', 'State', 'County/Parish'])
fixed_compls['missing_count'] = fixed_compls[measure_cols].isna().sum(axis=1)
fixed_compls = fixed_compls[fixed_compls['missing_count'] < 15].drop(columns='missing_count')

# Save to csv
fixed_compls.to_csv('fixed_compls.csv',index=False)
fixed_compls.head()
```

```python
join2=pd.merge(ascqr,fixed_compls,how='outer',on=['Facility Name', 'City/Town', 'State'])
join2.to_csv('join2.csv',index=False)
join2.head()
```

```python
# Repeat for fy_readmissions
fixed_fy_readmis = fy_readmissions.pivot_table(
    index=['Facility Name', 'State'],
    columns='Measure Name',
    values='Excess Readmission Ratio',
    aggfunc='first'  # use first score if duplicates exist
).reset_index()
fixed_compls.columns.name = None
fixed_fy_readmis.head()
```

```python
fy=pd.merge(fy_hac,fixed_fy_readmis,how='outer',on=['Facility Name', 'State'])
fy.head()
fy.to_csv('fy',index=False)
```

```python
join3=pd.merge(join2,fy,how='outer',on=['Facility Name','State'])
join3.to_csv('join3.csv',index=False)
join3.head()
```

```python
# Once again, we must repeat our data formating fixes
# Starting by formatting null values
hcahps['Patient Survey Star Rating'] = hcahps['Patient Survey Star Rating'].replace('Not Applicable', np.nan)
hcahps['HCAHPS Answer Percent']  = hcahps['HCAHPS Answer Percent'].replace('Not Applicable', np.nan)
# Combine value columns to one
hcahps['Value'] = hcahps['Patient Survey Star Rating'].combine_first(hcahps['HCAHPS Answer Percent'])
hcahps['Value'] = pd.to_numeric(hcahps['Value'], errors='coerce')
# Make pivot table
hcahps_pivot = hcahps.pivot_table(
    index=['Facility Name', 'City/Town', 'State', 'County/Parish'],
    columns='HCAHPS Answer Description',
    values='Value',
    aggfunc='first'    # if a facility/measure pair ever appears twice, just take the first
).reset_index()
hcahps_pivot.columns.name = None
hcahps_pivot.to_csv('hcahps_pivot.csv',index=False)
hcahps_pivot.head()
# 87 columns is too many I'll pull it into excel to examine and eliminate some

```

```python
# Pretty sure the only value column we need is summary star rating as it's a combination of all of them
hcahps_fixed=hcahps_pivot[['Facility Name','City/Town','State','Summary star rating']]
hcahps_fixed.to_csv('hcahps_fixed.csv', index=False)
hcahps_fixed.head()
```

```python
join4=pd.merge(join3,hcahps_fixed,how='outer',on=['Facility Name', 'City/Town','State'])
join4.to_csv('join4.csv',index=False)
join4.head()
# Starting to get a lot of null values to remove later...
```

```python
# Repeat for hai
hai_pivot = hai.pivot_table(
    index=['Facility Name', 'City/Town', 'State'],
    columns='Measure Name',
    values='Score',
    aggfunc='first'    # if a facility/measure pair ever appears twice, just take the first
).reset_index()
hai_pivot.columns.name = None
hai_pivot.to_csv('hai_pivot.csv',index=False)
hai_pivot.head()
```

```python
# Consolidated Data
hai_pivot=pd.read_csv('hai_pivot.csv')
hai_pivot.head()
```

```python
join5=pd.merge(join4,hai_pivot,how='outer',on=['Facility Name','City/Town','State'])
join5.to_csv('join5.csv',index=False)
join5.head()
```

```python
join6=pd.merge(join5,hgi,'outer',['Facility Name','City/Town','State'])
join6.to_csv('join6.csv',index=False)
join6.head()
```
